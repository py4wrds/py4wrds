# 0. Pre-course setup


We want to make sure everyone's ready to go before the first lesson begins!

This module will guide you through installing and configuring the two bits software needed for the workshop: VS Code and conda.

Please let us know if you encounter any issues.



## Install VS Code

VS Code (Visual Studio Code) is the development environment we'll be using for the course. It can edit Python files and notebooks, and comes with a number of tools to streamline Python software development.

Go to the [VS Code website](https://code.visualstudio.com/download) and download the installer for Windows. 

Double click the `.exe` file you downloaded to run the installer.

* You can accept all the default options in the install process.

Open the newly-installed "Visual Studio Code" from the Start menu. It should look something like this:

![extensions](img/0-vscode-welcome.png)




## Configure VS Code extensions

VS Code comes pre-installed with core functionalty, but there are a few extra extensions we'll be relying on for the course.

Open the extensions side panel using the extentions icon in the sidebar:

![extensions](img/0-vscode-extension-icon.png)

Search for and install the following extensions (all are authored by ![microsoft](img/0-vscode-ms-logo.png)):

* Python
* Jupyer
* Black Formatter (not the extension just called "Black"!)



## Install Conda

Conda (Anaconda) is a tool for managing Python dependencies. It was created to make it easy to install scientific computing Python packages.

Go to the [conda website](https://www.anaconda.com/download/success) and download the Windows installer.

Double click the `.exe` that was downloaded to run the installer.

* You can accept all the default options, including "Install for just me".
* Once conda in installed, you can exit without launching Anaconda Navigator: we won't be using it for this course. 

Ok, lets try it out! From the Start menu, launch **Anaconda PowerShell Prompt**.

* The logo looks like this: ![anaconda](img/0-anaconda-ps-icon.png)
* Conda also installs **Anaconda Prompt**. We won't be using that for this course, as the PowerShell prompt has more features and is more user friendly.

Inside the propt, type `conda --help` then press <kbd>ENTER</kbd>. You should see a list of commands like this, indicating a successful install of conda:

![conda](img/0-conda-ps-help.png)


## Create our conda environment

We put together a conda environment that contains all the Python packages and remaiing software needed for the workshop. You can also use this environment for your own projects!

Create a folder called `py4wrds`. You can do all course-releated development inside this folder.

Download this [environment.yml](https://github.com/py4wrds/py4wrds/raw/refs/heads/develop/environment.yml) file into your new `py4wrds` folder.

Open Anaconda prompt. Type `cd` followed by a space then the full path to your folder. Then press <kbd>ENTER</kbd> For me this looked like:

```bash
cd C:\\Users\Public\py4wrds
```



Type `conda create env --file environment.yml` and press <kbd>ENTER</kbd>. You should see a loading message like "Collecting packge metadata"



![conda](img/0-conda-env-create.png)



It can take over 30 minutes to install the full environment! Luckly this only has to be done once, and adding new packages in the future will be much faster. When it's done, you should see a `done` message:


![conda](img/0-conda-done.png)

From now on whenever we launch the conda prompt, you should run the command `conda activate py4wrds` to switch to your new environment with all your data science Python tools. After running this command, you should see the text at the start of each new line change from `(base)` to `(py4wrds)`. This indicates the active environment.


![conda](img/0-conda-activate.png)
