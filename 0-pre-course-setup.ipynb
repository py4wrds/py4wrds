## Install VS Code

VS Code (Visual Studio Code) is the development environment we'll be using for the course. It can edit Python files and notebooks, and comes with a number of tools to streamline Python software development.

Go do the VS Code website and download the installer for Windows. Double click the .msi file you downloaded to run the installer.

## Configure VS Code extensions

VS Code comes pre-installed with basic functionalty, but there are a few extra extensions we'll be relying on for the course.

Open the extensions side panel. Search and install both the "Jupyter" and "Python" extensions. The creater of both should be Microsoft.



## Install Conda

Conda (Anaconda) is a tool for managing Python dependencies. It was created to make it easy to install scientific computing Python packages.

Go to the conda website and download the Windows installer

* It's OK if the installer version is different to the one in the screenshot. Anaconda makes it easy to change Python versions.

Double click the .msi file.

Check options.




## Create our counda environment

We put together a conda environment that contains all the Python packages and remaiing software needed for the workshop. You can also use this environment for your own projects!

In VS Code, open the py4wrds folder.

Download this file into the folder. It should look like this.
